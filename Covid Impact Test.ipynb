```python
#IMPORTING DATASET(.CSV) INTO PANDAS DATAFRAME
import pandas as pd
dataset = pd.read_csv('02_May_12_covid_impact_survey.csv',
                       sep=',',
                       index_col=False,
                       dtype='unicode')
dataset
```

```python
sel_demog = ['GENDER','AGE4','RACETH','EDUCATION','HHINCOME','P_OCCUPY2','REGION4']
sel_social = ['SOC1','SOC2A','SOC2B','SOC3A','SOC3B','SOC4A','SOC4B']
sel_physical = ['PHYS8']
sel_mental = ['SOC5A','SOC5B','SOC5C','SOC5D','SOC5E']

sel_features = sel_demog + sel_social + sel_physical + sel_mental
sel_dataset = dataset[sel_features]
df=pd.get_dummies(sel_dataset,columns=sel_features)
df
```

```python
corrM = df.corr()
corrM
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# taking all rows but only 6 columns
sns.heatmap(corrM, annot = True)

plt.show
```
